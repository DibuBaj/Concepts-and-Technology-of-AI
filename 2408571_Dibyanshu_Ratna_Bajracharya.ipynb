**Task - 3 - Outlier Detection:**

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
data = df[df['Country name'].isin(south_asian_countries)]
df_quartiles = pd.DataFrame(data)
q1 = df_quartiles['score'].quantile(0.25)
q3 = df_quartiles['score'].quantile(0.75)
iqr = q3 - q1

q1_gdp = df_quartiles['Log GDP per capita'].quantile(0.25)
q3_gdp = df_quartiles['Log GDP per capita'].quantile(0.75)
iqr_gdp = q3_gdp - q1_gdp

print("1st quartile:", q1)
print("3rd quartile:", q3)
print("IQR:", iqr)
print("1st quartile GDP:", q1_gdp)
print("3rd quartile GDP:", q3_gdp)
print("IQR GDP:", iqr_gdp)
```

2. Define outliers using the 1.5 × IQR rule.

```python
outliers_score = data[(data['score'] < q1 - 1.5 * iqr) | (data['score'] > q3 + 1.5 * iqr)]
outliers_gdp = data[(data['Log GDP per capita'] < q1_gdp - 1.5 * iqr_gdp) | (data['Log GDP per capita'] > q3_gdp + 1.5 * iqr_gdp)]
outliers = pd.concat([outliers_score, outliers_gdp])
print("Outliers based on Score and GDP per Capita:")
outliers[['Country name','score','Log GDP per capita']]
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Log GDP per capita'], data['score'], color='orange', label='Regular Data')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='black', label='Outlier')
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.title('Scatter Plot of Score vs. GDP per Capita with Outliers')
plt.legend()
plt.show()
```

4. Discuss the characteristics of these outliers and their potential impact on regional averages.

Outlier Discussion:

In the provided data and graph, Afghanistan stands out as the only outlier among the South Asian countries. Even the next lowest Log GDP per capita score, recorded by Bangladesh at 1.122, is nearly double that of Afghanistan, highlighting a stark disparity.

Why is Afghanistan an Outlier?

Using the 1.5 × IQR Rule:

For both the happiness score and Log GDP per capita, Afghanistan’s values fall far below the lower bounds, calculated as:

Lower Bound = Q1 − 1.5 × IQR

This confirms Afghanistan’s position as an outlier in the dataset.

Impact of Afghanistan as an Outlier:

Afghanistan’s extreme values have the potential to skew the regional averages for South Asia. If included without adjustments, it could significantly lower the overall mean happiness scores and related metrics, thereby distorting the representation of the region’s performance. To better understand the typical trends in South Asia, statistical analyses may need to either account for or exclude Afghanistan from the calculations.
